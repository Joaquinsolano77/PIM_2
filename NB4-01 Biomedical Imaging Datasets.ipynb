## 1. Introduction

A dataset for deep learning typically refers to a *`specialized repository`* of data structured and optimized for training and evaluating deep learning models. These dataset `differ from traditional datasets` in several key ways:

1. **`Large-scale and Diverse Data`**: Deep learning datasets often contain *`massive amounts of data`*, which is crucial for training models with high accuracy. This data is *`usually diverse`*, encompassing various types such as images, text, audio, or video.

2. **`Annotated and Labeled Data`**: Data in deep learning datasets is meticulously *`annotated and labeled`*. Annotations provide *ground truth* information essential for supervised learning tasks, enabling models to learn from labeled examples.

3. **`High Dimensionality and Complexity`**: Deep learning datasets can have *`high dimensionality`*, especially in the case of images and videos where each *data point (e.g., pixel or frame) contributes* to the overall *`complexity`* of the dataset.

4. **`Optimized for Performance`**: These datasets are optimized for performance to handle large-scale data processing efficiently. They often leverage *`distributed computing frameworks`* and *`storage systems`* to manage and process data in parallel.

5. **`Version Control and Management`**: Due to the dynamic nature of data in deep learning research, *`version control and management of datasets are critical`*. Tools and practices are employed *`to track changes, ensure data integrity, and facilitate reproducibility in experiments`*.

6. **`Privacy and Security Considerations`**: Deep learning datasets may contain *`sensitive information`*, necessitating robust privacy and security measures to protect data confidentiality and prevent unauthorized access.

7. **`Community and Accessibility`**: Many deep learning datasets are *`open-source or publicly accessible`*, fostering collaboration and enabling researchers worldwide to contribute, use, and build upon existing datasets.

In essence, _`a database for deep learning serves as a foundational resource for training, validating, and benchmarking deep learning models`_. Its specialized characteristics cater to the unique requirements of machine learning algorithms, facilitating advancements in artificial intelligence research and applications.



### Public Datasets

There are several public medical imaging datasets that are available to the research community and the general public. _`Some of the most well-known public medical imaging datasets include`_:

This dataset spans _`1000 object classes and contains 1,281,167 training images, 50,000 validation images and 100,000 test images`_. This subset is available on [Kaggle](https://www.kaggle.com/c/imagenet-object-localization-challenge/overview/description).


4. [**Open Access Biomedical Image Search Engine (Openi):**](https://libraries.usc.edu/databases/open-i-open-access-biomedical-image-search-engine) Openi is a _`search tool that allows you to access a wide variety of medical images and their metadata`_. You can search for images based on specific medical keywords.

6. [**Radiological Society of North America (RSNA) Radiology Cases Database:**](https://cases.rsna.org/) This dataset contains a _`collection of radiological cases`_, including X-ray, CT, MRI and ultrasound images.

### What Is a JSON File?

A **``JSON file``** (JavaScript Object Notation) is a lightweight text-based format _`used to store and transmit structured data`_. JSON is _``easy for humans``_ to read and write, and it's _``easy for machines``_ to parse and generate. It's widely used in web applications, APIs, and databases because of its simplicity and compatibility with most programming languages.

#### Structure of a JSON file:
A JSON file contains data organized as _``key-value pairs``_ and uses a syntax similar to JavaScript objects. It is based on _``two primary structures``_:

1. **``Objects``**: Collections of key-value pairs, enclosed in curly braces `{}`.
2. **``Arrays``**: Ordered lists of values, enclosed in square brackets `[]`.

Example of a JSON file:
```json
{
  "name": "John",
  "age": 30,
  "married": false,
  "children": ["Anna", "Peter"],
  "address": {
    "street": "Fake Street",
    "city": "Springfield"
  }
}
```
>**[COCO JSON Format](https://roboflow.com/formats/coco-json)**
>
>The `COCO JSON format` is a file _``format used to store annotations for datasets``_ in computer vision tasks, particularly object detection, instance segmentation, and related tasks. `COCO (Common Objects in Context)` is a widely used standard in the computer vision community and provides a `structured framework for organizing and labeling image data`.
>
>_``The COCO JSON file consists of several main components``_:
>
>1. **`Info`**: General information about the dataset, such as the year, version, description, and contributor.
    ```json
    {
      "year": 2024,
      "version": "1.0",
      "description": "Example Dataset",
      "contributor": "Example",
      "date_created": "2024-07-06"
    }
    ```
>
>2. **`Licenses`**: Information about the licenses of the images contained in the dataset.
    ```json
    [
      {
        "id": 1,
        "name": "Example License",
        "url": "http://example.com/license"
      }
    ]
    ```
>
>3. **`Images`**: Information about each image in the dataset, including the image ID, file name, height, and width.
    ```json
    [
      {
        "id": 1,
        "width": 640,
        "height": 480,
        "file_name": "image1.jpg",
        "license": 1,
        "date_captured": "2024-07-06"
      }
    ]
    ```
>
>4. **`Annotations`**: Annotations associated with the images, including information about object categories, object locations (bounding boxes), and segmentation masks.
    ```json
    [
      {
        "id": 1,
        "image_id": 1,
        "category_id": 1,
        "segmentation": [[100.0, 100.0, 150.0, 100.0, 150.0, 150.0, 100.0, 150.0]],
        "area": 2500.0,
        "bbox": [100.0, 100.0, 50.0, 50.0],
        "iscrowd": 0
      }
    ]
    ```
>
>5. **`Categories`**: Information about the different object categories present in the dataset.
    ```json
    [
      {
        "id": 1,
        "name": "person",
        "supercategory": "human"
      }
    ]
    ```
>
>These components are structured in a JSON file that can be processed by various libraries and tools for computer vision tasks. This format facilitates interoperability and the use of datasets in various projects and experiments.

JSON's simplicity and flexibility make it a ``popular choice for data exchange`` in web and software development.

````python
import os
import json
import subprocess

def configure_kaggle_api():
    try:
        # Verificar si el archivo kaggle.json existe
        if not os.path.exists('./kaggle.json'):
            raise FileNotFoundError("El archivo kaggle.json no se encontró en el directorio actual.")

        # Verificar que el archivo kaggle.json tenga un formato JSON válido
        with open('./kaggle.json', 'r') as f:
            try:
                json.load(f)
            except json.JSONDecodeError:
                raise ValueError("El archivo kaggle.json no tiene un formato JSON válido.")

        # Crear el directorio ~/.kaggle si no existe
        subprocess.run(['mkdir', '-p', '~/.kaggle'], check=True)
        print("Directorio ~/.kaggle creado o ya existente.")

        # Copiar el archivo kaggle.json al directorio ~/.kaggle
        result = subprocess.run(['cp', './kaggle.json', '~/.kaggle/'], check=True)
        if result.returncode == 0:
            print("Archivo kaggle.json copiado exitosamente.")
        else:
            print("Error al copiar el archivo kaggle.json.")

        # Cambiar permisos para el archivo kaggle.json
        result = subprocess.run(['chmod', '600', '~/.kaggle/kaggle.json'], check=True)
        if result.returncode == 0:
            print("Permisos cambiados exitosamente para kaggle.json.")
        else:
            print("Error al cambiar permisos para kaggle.json.")

        # Listar el contenido del directorio de inicio para verificar
        subprocess.run(['ls', '-al', '~'], check=True)

    except FileNotFoundError as fnf_error:
        print("Error:", fnf_error)
    except ValueError as ve:
        print("Error de formato:", ve)
    except subprocess.CalledProcessError as e:
        print("Error en la ejecución del comando:", e)
    except Exception as e:
        print("Ocurrió un error inesperado:", e)

# Ejecutar la función
configure_kaggle_api()


#Este código importa las librerias necesarias para luego configura el entorno para usar la API de
#Kaggle al crear la carpeta .kaggle, copia el archivo
#kaggle.json, ajusta permisos lo verifica y comprueba el
#contenido, el formato, y dandonos los posibles errores que pueden aparecer con lo mencionado
````

### [Leukaemia](https://www.cancer.org/cancer/types/leukemia.html)

A brief internet search indicates the following about Leukaemia.

**`Leukaemia`**, which is a type of *`blood cancer`*, is a group of diseases that affect the blood cells and bone marrow. There are several types of leukaemia, the two main ones being *`acute myeloid leukaemia`* (AML) and *`acute lymphoblastic leukaemia`* (ALL). In addition to these, there are other subtypes of leukaemia, including *`chronic myeloid leukaemia`* (CML) and *`chronic lymphocytic leukaemia`* (CLL).

Main Types of Leukemia:

1. **`Acute Myeloid Leukaemia (AML)`:**
   - It is a type of leukaemia that affects immature myeloid cells.
   - It can progress rapidly if left untreated.

2. **`Acute Lymphoblastic Leukaemia (ALL)`:**
   - It affects immature lymphoid cells.
   - It is more common in children but can also affect adults.

3. **`Chronic Myeloid Leukaemia (CML)`:**
   - It is characterized by uncontrolled growth of myeloid cells.
   - It tends to progress more slowly.

4. **`Chronic Lymphocytic Leukaemia (CLL)`:**
   - It involves excessive growth of mature lymphocytic cells.
   - It is more common in adults and generally progresses slowly.

### Causes of Leukaemia:

The exact causes of leukemia are not always clear, but some risk factors and potential causes include:

1. **`Genetics`:**
   - Some forms of leukaemia have a genetic component, meaning there may be a hereditary predisposition.

2. **`Exposure to Radiation`:**
   - Exposure to high levels of ionizing radiation can increase the risk of leukaemia.

3. **`Exposure to Chemicals`:**
   - Prolonged exposure to certain chemicals, such as benzene, can increase the risk.

4. **`Genetic Disorders`:**
   - Some genetic disorders, such as Down syndrome, can increase the risk of leukemia.

5. **`Medical Treatments`:**
   - Radiation therapy and certain chemotherapy treatments for other types of cancer can increase the risk of developing leukaemia.

It is important to note that while these factors may increase the risk, they do not guarantee that a person will develop leukaemia. The disease is complex and can affect individuals without known risk factors.

## 4. Determining Physical Characteristics of Cells with Cancer



---
**Uses OpenCV to determine area, perimeter (contour), and moments (centroid)**

---


This code performs the following steps:

1. **`Choosing a random image`:**
   - Randomly selects an image from the specified directory.

2. **`Reading and thresholding the image`:**
   - Reads the image in grayscale.
   - Applies inverse binary thresholding.

3. **`Finding and sorting contours`:**
   - Finds contours in the thresholded image.
   - Prints the number of objects found.
   - Sorts the contours by area from largest to smallest.

4. **`Calculating features`:**
   - Calculates the area, perimeter, centroid, and bounding box of the largest contour.

5. **`Visualization with subplots`:**
   - Sets up subplots to display the original image, the thresholded image, the contour with centroid and bounding box, and finally, the image with contours and bounding box drawn using `cv2`.

6. **`Printing features`:**
   - Prints the area, perimeter, and centroid of the largest contour.


````python
# CÓDIGO MEJORADO

import cv2
import numpy as np
import random
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import os

# Elige una imagen aleatoria de la lista 'image_files'
random_image = random.choice(image_files)

# Comprobación de errores: Verificar que el archivo de la imagen existe en el directorio
random_image_path = os.path.join(directory, random_image)
if not os.path.exists(random_image_path):
    raise FileNotFoundError(f"La imagen {random_image} no se encuentra en el directorio {directory}.")

# Leer la imagen en formato escala de grises (blanco y negro)
image = cv2.imread(random_image_path, cv2.IMREAD_GRAYSCALE)

# Comprobación de errores: Verificar que la imagen se haya cargado correctamente
if image is None:
    raise ValueError(f"No se pudo leer la imagen desde {random_image_path}. Asegúrate de que el archivo esté en formato correcto.")

# Aplicar un umbral para convertir la imagen en blanco y negro
# El umbral 128 convierte los píxeles de 0-127 a negro (0) y los de 128-255 a blanco (255)
_, thresholded = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY_INV)

# Encontrar los contornos de la imagen umbralizada
# Se usan los contornos más externos y se simplifican
contours, hierarchy = cv2.findContours(thresholded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Comprobación de errores: Verificar que se hayan encontrado contornos
if len(contours) == 0:
    raise ValueError("No se encontraron contornos en la imagen después de aplicar el umbral.")

# Imprimir la cantidad de objetos (contornos) encontrados
num_objects = len(contours)
print(f'Número de objetos encontrados: {num_objects}')

# Ordenar los contornos por área, de mayor a menor
contours = sorted(contours, key=cv2.contourArea, reverse=True)

# Calcular las características del contorno más grande
area = cv2.contourArea(contours[0])
# Verificación de errores: Comprobar que el área no sea cero (lo que indicaría un contorno inválido)
if area == 0:
    raise ValueError("El área calculada es 0, lo que indica un contorno inválido.")

# Calcular el perímetro del contorno más grande
perimeter = cv2.arcLength(contours[0], True)

# Calcular los momentos para obtener el centroide del contorno más grande
moments = cv2.moments(contours[0])
if moments["m00"] == 0:
    raise ValueError("Los momentos del contorno son inválidos (división por cero).")
centroid_x = int(moments["m10"] / moments["m00"])
centroid_y = int(moments["m01"] / moments["m00"])

# Obtener la caja delimitadora (bounding box) del contorno más grande
x, y, w, h = cv2.boundingRect(contours[0])

# Crear subgráficas para visualizar los resultados
fig, axs = plt.subplots(1, 4, figsize=(12, 4))

# Mostrar la imagen original
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Imagen Original')
axs[0].axis('off')  # Ocultar los ejes para mayor claridad

# Mostrar la imagen umbralizada (blanco y negro)
axs[1].imshow(thresholded, cmap='gray')
axs[1].set_title('Umbralizado')
axs[1].axis('off')

# Mostrar el contorno más grande y la caja delimitadora sobre la imagen original
# Convertimos la imagen a color para poder mostrar el contorno y la caja en colores visibles
image_bgr = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
axs[2].imshow(image_bgr)
axs[2].plot(
    np.append(contours[0][:, 0, 0], contours[0][0, 0, 0]),  # Coordenadas X del contorno
    np.append(contours[0][:, 0, 1], contours[0][0, 0, 1]),  # Coordenadas Y del contorno
    color='lime', linewidth=2  # Color verde lima para el contorno
)

# Agregar la caja delimitadora y el centroide en la visualización
rect = Rectangle((x, y), w, h, linewidth=2, edgecolor='red', facecolor='none')
axs[2].add_patch(rect)  # Dibujar el rectángulo de la caja delimitadora
axs[2].scatter(centroid_x, centroid_y, color='red', s=100, marker='x')  # Dibujar el centroide
axs[2].set_title('Contorno, Centroide y BBox')

# Mostrar los contornos y la caja delimitadora utilizando la función drawContours de OpenCV
cv2.drawContours(image_bgr, [contours[0]], -1, (0, 255, 0), 1)  # Contorno en verde
cv2.rectangle(image_bgr, (x, y), (x + w, y + h), (0, 0, 255), 1)  # Caja delimitadora en rojo
axs[3].imshow(image_bgr)
axs[3].set_title('Contornos y BBox (cv2)')
axs[3].axis('off')  # Ocultar los ejes

# Ajustar el layout de las subgráficas para evitar solapamientos
plt.tight_layout()

# Mostrar la visualización
plt.show()

# Imprimir las características calculadas
print(f"Área: {area}")
print(f"Perímetro: {perimeter}")
print(f"Centroide: ({centroid_x}, {centroid_y})")
````

Apart from **`area, perimeter, and centroid`**, there are several other features that can be extracted or shown from the contour of an object in an image. Some of these features include:

1. **`Compactness`**: Measure of how compact the shape is, calculated as the square of perimeter divided by area. Higher values indicate more irregular shapes.

2. **`Elongation`**: Measure of how elongated or stretched the shape is, calculated as the ratio of the major axis length to the minor axis length of the fitted ellipse.

3. **`Extent`**: Ratio of the contour area to the bounding box area. It measures how much of the bounding box is filled by the object.

4. **`Solidity`**: Ratio of contour area to convex hull area. It measures the proportion of the convex hull that is filled by the object.

5. **`Eccentricity`**: Measure of how much the shape deviates from a perfect circle. It is calculated as the square root of the ratio of the major and minor axes of the fitted ellipse.

6. **`Orientation`**: Angle of the major axis of the fitted ellipse with respect to the x-axis.

7. **`Convexity Defects`**: Points where the contour deviates significantly from the convex hull. These points can indicate concavities or holes in the object.

8. **`Hu Moments`**: A set of seven moments invariant to translation, rotation, and scale. They capture the shape of the object more robustly than raw moments.

9. **`Haralick Texture Features`**: Statistical measures derived from the gray-level co-occurrence matrix (GLCM), describing texture characteristics such as contrast, energy, and entropy.

These features can provide valuable information about the shape, texture, and overall characteristics of objects in an image, which can be useful for various applications such as **`object recognition, classification, and segmentation`**.
